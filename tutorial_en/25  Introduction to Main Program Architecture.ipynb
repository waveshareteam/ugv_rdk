## Installation Script

In the project folder, there is a file named setup.sh, written in shell script, which helps automate the configuration of the upper computer for the robot product. This includes setting up the serial port, configuring the camera, creating a project virtual environment, and installing dependencies. These steps are already configured in the SD card image we provide.

Usage of the installation script: The installation process involves downloading and installing many dependencies from the internet. For areas with special network environments, we recommend downloading the image file directly from our official website to install the product.

## Automatic Program Execution

The autorun.sh script in the project folder is used to configure the automatic startup (as a user, not root) of the main program (app.py) and JupyterLab (start_jupyter.sh), and generates the configuration file for JupyterLab.terLab.

## Installation Script
In the project folder, there is a file named setup.sh, written in shell script, which helps automate the configuration of the upper computer of the robot product, including setting up serial ports, configuring cameras, creating project virtual environments, and installing dependencies. These steps are already configured in the image of our factory SD card.  

Usage of the installation script: The installation process requires downloading and installing many dependencies from the network. For regions with special network environments, we recommend using the method of directly downloading the image file from our official website to install the product.  

The setup.sh script needs to be run with root privileges.  

> cd ugv_rpi/  
> sudo chmod +x setup.sh  
> sudo ./setup.sh

## Automatic Program Execution
The autorun.sh file in the project folder is used to configure the automatic startup of the main program (app.py) and jupyterlab (start_jupyter.sh) of the product (as a user rather than root), and generate the configuration file for jupyterlab.  

The autorun.sh script needs to be run with user privileges.  

> sudo chmod +x autorun.sh  
> ./autorun.sh

## Updating the Project
This project will be continuously updated. When you need to update, simply navigate to the project folder and use the git pull command to update. Note that this operation will replace any changes you have made before. After the update, you need to change the parameters in config.yaml to set the chassis type and module type. You can also use the command-line tool of the WEB control interface to configure the product type.  

> cd ugv_rpi/  
> git pull

After restarting the device, open the WEB control interface and enter the command to set the product type, for example:

> s 20

The first digit 2 represents the chassis type, and the second digit 0 represents no module or PT module.  

The first digit:

- 1: RaspRover
- 2: UGV Rover
- 3: UGV Beast  

The second digit:

- 0: None/PT
- 1: RoArm
- 2: Pan-Tilt

This configuration only needs to be set once after the update or reinstallation and will be saved in config.yaml.
